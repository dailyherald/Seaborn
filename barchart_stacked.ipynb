```python
import pandas as pd
import matplotlib as mpl
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
%matplotlib inline

mpl.rcParams['font.family'] = ['sans-serif']
mpl.rcParams['xtick.labelsize'] = 7
mpl.rcParams['ytick.labelsize'] = 7
mpl.rcParams['pdf.fonttype'] = 42 # allegedly allows text to be saved as editable

mpl.rcParams['font.sans-serif'] = ['Arial Narrow']
mpl.rcParams['text.usetex'] = False
mpl.rcParams['svg.fonttype'] = 'none'

import warnings
warnings.filterwarnings("ignore")

print('ready')
```

```python
df=pd.read_csv('ElginUOF.csv')
df.head(3)
```

```python
df['yearQuarter'] = df['year'].astype(str) + ', ' + df['quarter']
df.head(3)
```

```python
df.dtypes
```

```python
# bar chart with white text labels

my_dpi=150

f, ax = plt.subplots(1,figsize=(600/my_dpi, 400/my_dpi), dpi=my_dpi)

#plt.xlim(0, 40)

addup = df['Show of force Only'] + df['Total Transitions (show to use of force)']

# Stacked charts get complicated
# so for bottom one, use total
# for the middle one, add that and the top one
# then just the top one
g1 = ( sns.barplot(
        x="Total Response to Resistance (RTR) Incidents", 
        y="yearQuarter", 
        data=df,
        label="UOF only",
        color="LightBlue"
        ),
        sns.barplot(
        x= addup,
        y='yearQuarter',
        data=df,
        label="SOF only",
        color="DeepSkyBlue"
        ),
        sns.barplot(
        x='Total Transitions (show to use of force)',
        y='yearQuarter',
        data=df,
        label="Transition",
        color="DarkBlue")
    )



# Add a legend
ax.legend(ncol=1, loc="upper right", frameon=False)

f.suptitle('Elgin police calls', fontsize=14, family='Arial Black', y=1.15)
ax.set_title('By quarter, 2014-\'16 but let\'s have some more \n text to see what happens when this goes \n over multiple lines', fontsize=10)#, loc='left')
ax.set_ylabel('YEAR, QUARTER', fontsize=9, fontweight='bold')
ax.set_xlabel('TOTAL FORCE INCIDENTS', fontsize=9, fontweight='bold')
f.text(0,0-.05, 'SOURCE: ELGIN POLICE | DAILY HERALD GRAPHIC', fontsize=6)

# No bar labels

f.savefig('pdfs_images/stacked_barchart.pdf')
```
